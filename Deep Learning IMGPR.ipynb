# Image Processing Technique
### 1. RGB Gray Scale
#### Gray Scale:- Convert the image into black and white image

```python
import cv2

img = cv2.imread('Dog.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2. Bluring
#### Bluring:- Blurs the complete image

```python
import cv2

img = cv2.imread('Dog.jpg')
blur = cv2.GaussianBlur(img, (15, 15), 0)

cv2.imshow('Blurred Image', blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 3. Edge Detection
#### Edge Detection : - Only gives the edge of the image object present

```python
import cv2

img = cv2.imread('Dog.jpg', 0)
edges = cv2.Canny(img, 100, 200)

cv2.imshow('Edges', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4. Threshold
#### Threshold: - convert the image in dot image form in black and white colour

```python
import cv2

img = cv2.imread('Dog.jpg', 0)
_, threshold = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

cv2.imshow('Threshold Image', threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 5. Dilation and Erosion
#### Dilation:- Incresing the object boundaries
#### Erosion:- Decreasing the object boundaries

```python
import cv2
import numpy as np

img = cv2.imread('Dog.jpg', 0)
kernel = np.ones((5,5), np.uint8)

dilation = cv2.dilate(img, kernel, iterations=1)
erosion = cv2.erode(img, kernel, iterations=1)

cv2.imshow('Dilated Image', dilation)
cv2.imshow('Eroded Image', erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 6. Cropping the Image
#### Crop the image 

```python
import cv2

img = cv2.imread('Dog.jpg')
cropped = img[50:200, 100:300]

cv2.imshow('Cropped Image', cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 7. Image Resize
#### Resize:- Resize the image according to our convinious

```python
import cv2

img = cv2.imread('Dog.jpg')
resized = cv2.resize(img, (300, 300))

cv2.imshow('Resized Image', resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 8. HSV(Hue staturation and value)
#### It is a Filter for applying colour

```python
import cv2

img = cv2.imread('Dog.jpg')
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

cv2.imshow('HSV Image', hsv_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 9. Image Rotation 

```python
import cv2

img = cv2.imread('Dog.jpg')
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

angle = 45     #If we give the angle in negative then it we move anti clock wise
scale = 1.0

rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated_img = cv2.warpAffine(img, rotation_matrix, (w, h))        #WarpAffine-help in applying the tranformation

cv2.imshow('Rotated Image', rotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 10. Morphological Transformation

```python
import cv2
import numpy as np

img = cv2.imread('Dog.jpg', 0)
kernel = np.ones((5,5), np.uint8)

opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

cv2.imshow('Opening (Noise Removal)', opening)
cv2.imshow('Closing (Fill Gaps)', closing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 11. Contouring the image
#### Analysing the shape of an object
#### Finding the join or connection in the image

```python
import cv2

img = cv2.imread('Dog.jpg', 0)
_, threshold = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(img, contours, -1, (0, 255, 0), 3)

cv2.imshow('Contours', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 12. Histogram
#### Improve or enhance the contrast of the image

```python
import cv2

img = cv2.imread('Dog.jpg', 0)
equalized = cv2.equalizeHist(img)

cv2.imshow('Equalized Image', equalized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 13. Template Matching
#### In large image it finds the similarity

```python
import cv2

img = cv2.imread('Dog.jpg',0)
template = cv2.imread('template.jpg', 0)
result = cv2.matchTemplate(img, template, cv2.TM_CCOEFF_NORMED)

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
top_left = max_loc
h, w = template.shape[:2]
bottom_right = (top_left[0] + w, top_left[1] + h)

cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 3)

cv2.imshow('Detected Template', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Concatenation Image 

```python
import cv2
import numpy as np

img1 = cv2.imread('Dog.jpg')
img2 = cv2.imread('House.jpg')

img1 = cv2.resize(img1, (500, 500))
img2 = cv2.resize(img2, (500, 500))

h_concat = np.hstack((img1, img2))
v_concat = np.vstack((img1, img2))

cv2.imshow('Horizontal Concatenation', h_concat)
cv2.imshow('Vertical Concatenation', v_concat)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
